```python
import pandas as pd
import json
UniList=pd.read_csv('UniDetails.csv')
UniCourse=pd.read_csv('Courses.csv')
finaldict={}
user_input=input()
match=UniList[UniList["Institution Name"].str.contains(user_input,case=False,na=False)]
for i,rows in match.iterrows():
    name=rows['Institution Name']
    courses=UniCourse[UniCourse['Institution Name'].str.contains(name,case=False,na=False)].to_dict(orient='records')
    finaldict[rows['Institution Name']]={
        'University_Detail':rows,
        'Coursedetail':courses
    }
print(finaldict)
```
